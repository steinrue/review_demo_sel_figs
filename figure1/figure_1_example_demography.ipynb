```python
# import modules
import demesdraw
import stdpopsim
import matplotlib.pyplot as plt
```

```python
# two plots next to each other
(fig, ax) = plt.subplots(1, 2, figsize=(8,4))

# Draw demographic model from Browning et. al. (2018) on left
species = stdpopsim.get_species("HomSap")
model = species.get_demographic_model("AmericanAdmixture_4B11")
graph = model.model.to_demes()
leftAx = demesdraw.tubes (graph, ax=ax[0], log_time=True)

# Draw demographic model from Browning et. al. (2018) on left
species = stdpopsim.get_species("HomSap")
model = species.get_demographic_model("AmericanAdmixture_4B11")
graph = model.model.to_demes()
leftAx = demesdraw.tubes (graph, ax=ax[0], log_time=True)
tmp = leftAx.set_title ("Browning et. al. (2018)")
leftAx.set_xticklabels (["YRI", "CEU", "CHB/JPT", "ADMIX"])

# Draw demographic model from Kuhlwilm et. al. (2019) on left
species = stdpopsim.get_species("PanTro")
model = species.get_demographic_model("BonoboGhost_4K19")
graph = model.model.to_demes()
rightAx = demesdraw.tubes (graph, ax=ax[1], log_time=True)
tmp = rightAx.set_title ("Kuhlwilm et. al. (2019)")

fig.tight_layout()
fig.savefig ("figure_1_example_demography.pdf")
```
